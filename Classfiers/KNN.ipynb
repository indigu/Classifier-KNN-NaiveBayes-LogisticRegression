```python
import pandas as pd
import numpy as np
import math
import operator
import sklearn
```

```python
# Importing data 
feature_names=['id',
'age',
'job',
'marital',
'education',
'default',
'balance',
'housing',
'loan',
'contact',
'day',
'month',
'duration',
'campaign',
'pdays',
'previous',
'poutcome',
'y']

data = pd.read_csv('trainingset.txt', delimiter=",", names=feature_names)

#data = data[data.pdays != -1]
#data = data[data.contact != "unknown"]
#data = data.drop('id', axis='columns')

from sklearn import preprocessing
le = preprocessing.LabelEncoder()

data = data.apply(le.fit_transform)
data.head()
```

```python
dataset = data[['id',
'age',
'job',
'marital',
'education',
'default',
'balance',
'housing',
'loan',
'contact',
'day',
'month',
'duration',
'campaign',
'pdays',
'previous',
'poutcome']].copy()

target_name = data[['y']].copy()
```

```python
dataset.head()
```

```python
# shows how uneven our dataset is
target_name['y'].value_counts()
```

```python
print(dataset.shape)
```

```python
print(target_name.shape)
```

```python
from sklearn import preprocessing
names = dataset.columns
# creating scaler object
scaler = preprocessing.StandardScaler()
scaled_dataset = scaler.fit_transform(dataset)
scaled_dataset = pd.DataFrame(scaled_dataset, columns=names)
scaled_dataset.shape
```

```python
# shows how uneven our dataset is
target_name['y'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
# Standardized Data
# Splitting Data into 80/20 
X_train, X_test, y_train, y_test = train_test_split(scaled_dataset, target_name, test_size=0.2,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
#Create KNN Classifier
knn = KNeighborsClassifier(n_neighbors=5)

#Train the model using the training sets
knn.fit(X_train, y_train.values.ravel())

#Predict the response for test dataset
y_pred = knn.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test,y_pred))
```
